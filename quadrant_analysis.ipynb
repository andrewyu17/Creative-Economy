```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
```

## Load Data

```python
df = pd.read_stata("/Users/andrewyu/Documents/Creative Economy/Data/quad_analysis.dta")
df.head()
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=df, x='changeE1519', y='changeE1921')

plt.title('Pre-Pandemic Employment Change vs Post-Pandemic Employment Change')
plt.xlabel('Employment Change (2015-2019)')
plt.ylabel('Employment Change (2019-2021)')

# Scatter labels
texts = [plt.text(x=df.changeE1519[i], 
                  y=df.changeE1921[i], 
                  s=df.subsector[i], size=9) for i in range(df.shape[0])]
adjust_text(texts, arrowprops=dict(arrowstyle='-', connectionstyle='arc3,rad=0.', color='red', lw=1))

# Axis and limits          
plt.axhline(y=0, color='red', linestyle='--', linewidth=1, alpha=0.4)           
plt.axvline(x=0, color='red',linestyle='--', linewidth=1, alpha=0.4)
plt.xlim([-30, 55])
plt.ylim([-45, 15])

# Quadrant Markers
plt.text(x=40, y=-35, s="Disrupted",alpha=0.7,fontsize=14, color='b')
plt.text(x=-25, y=-35, s="Already Declining",alpha=0.7,fontsize=14, color='b')
plt.text(x=-25, y=5, s="Anti-Fragile", alpha=0.7,fontsize=14, color='b')
plt.text(x=40, y=5, s="Resilient", alpha=0.7,fontsize=14, color='b')
plt.tight_layout()
plt.savefig("/Users/andrewyu/Documents/Creative Economy/Graphs/quad_analysis/emplt_change.png")
```

```python
counties = pd.read_stata("/Users/andrewyu/Documents/Creative Economy/Data/quad_analysis_counties.dta")
counties
```

```python
design = counties.loc[counties['category'] == "Design"]
whole = counties.loc[counties['category'] == "Wholesaling"]
manu = counties.loc[counties['category'] == "Manufacturing"]
mp = counties.loc[counties['category'] == "Media & Publishing"]
ae = counties.loc[counties['category'] == "Arts & Entertainment"]
```

```python
mp
```

```python
g = sns.FacetGrid(mp, 
                  col = 'area', 
                  col_wrap = 3,
                  hue='subsector',
                  col_order=['Los Angeles County','Orange County','San Bernardino County','Riverside County','Alameda County','Contra Costa County','San Francisco County','Santa Clara County','San Diego County'])
g.map(sns.scatterplot,'changeE1519','changeE1921').add_legend(title="Subsectors in Media & Publishing")
g.set_axis_labels('Employment Change (2015-2019)', 'Employment Change (2019-2021)', size = 8)
g.set_titles(col_template="{col_name}", size=10)
g.set(xlim=(-100,100),ylim=(-100,100))
g.fig.suptitle("Employment Change in Media & Publishing Pre-Pandemic vs Post-Pandemic", y=1) 
for a in g.axes:
    a.axhline(0, alpha=.5, color='red',lw=0.5,linestyle='--')
    a.axvline(0, alpha=.5, color='red',lw=0.5,linestyle='--')
for axis in g.axes.flat:
    axis.tick_params(labelleft=True, labelbottom=True, labelsize=5)
    axis.tick_params(labelbottom=True)
    axis.text(50, 80,'Resilient', fontsize=7)
    axis.text(-90, 80,'Anti-Fragile', fontsize=7)
    axis.text(50, -85,'Disrupted', fontsize=7)
    axis.text(-90, -85,'Already \nDeclining', fontsize=7)

g.tight_layout()
plt.savefig("/Users/andrewyu/Documents/Creative Economy/Graphs/quad_analysis/cempltMP.png")
```
